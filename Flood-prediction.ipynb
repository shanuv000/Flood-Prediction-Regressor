## Fecth Data

```python

def load_data(csv_name, csv_url):
    if os.path.isfile(f'{csv_name}.csv'):
        # If the file exists, read it
        dataset = pd.read_csv(f'{csv_name}.csv')
    else:
        # If the file doesn't exist, download it and read it
        dataset = pd.read_csv(csv_url)
        # Save the downloaded file for future use
        dataset.to_csv(f'{csv_name}.csv')  # Index is set to False to avoid saving index as a column

    return dataset
```

```python
train_url = 'https://raw.githubusercontent.com/shanuv000/csv_data/main/Flood%20prediction%20Dataset/train.csv'
test_url = 'https://raw.githubusercontent.com/shanuv000/csv_data/main/Flood%20prediction%20Dataset/test.csv'
sub_url = 'https://raw.githubusercontent.com/shanuv000/csv_data/main/Flood%20prediction%20Dataset/sample_submission.csv'
train_data = load_data('train', train_url)
test_data = load_data('test', test_url)
sub_data = load_data('sample_submission',sub_url)
```

```python
# Drop unnecesarry columns
columns = 'id'
train_data = train_data.drop(columns=columns)
test_data = test_data.drop(columns = columns)
```

```python
# X,y preparation
target_cols = 'FloodProbability'
X_train = train_data.drop(target_cols,axis=1)
y_train = train_data[target_cols]
X_test = test_data.copy()
```

```python
def add_features(df):
    
    """
    Add various statistical features to pandas DataFrame.
    
    Parameters:
    pandas DataFrame: Input DataFrame containing numerical data.
    
    Returns:
    pandas DataFrame: DataFrame with additional statistical features added.
    """
    features = df.select_dtypes(include = np.number).columns.to_list()

    df['mean_features'] = df[features].mean(axis=1)
    df['std_features'] = df[features].std(axis=1)
    df['max_features'] = df[features].max(axis=1)
    df['min_features'] = df[features].min(axis=1)
    df['median_features'] = df[features].median(axis=1)
    df['range_features'] = df['max_features'] - df['min_features']
    df['variance_features'] = df[features].var(axis=1)
    df['skewness_features'] = df[features].skew(axis=1)
    df['kurtosis_features'] = df[features].kurtosis(axis=1)
    df['sum_features'] = df[features].sum(axis=1)
    return df
    
```

```python
X_train = add_features(X_train)
X_test = add_features(X_test)
```

```python
X_train.describe().T
```

## Linear regression

```python
# 5 Fold evaluation with StandardScaler

steps = [('scaler', StandardScaler()),('model',LinearRegression())]
pipeline = Pipeline(steps=steps)
n_splits = 5
cv = KFold(n_splits=n_splits, shuffle=True)
scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='r2')

# Result
print(f"Mean(Linear Regression): {np.mean(scores)}")
print(f"std(Linear Regression): {np.std(scores)}")
```

## LGBM Regressor

```python
# 5 Fold evalutation with StandardScaler
steps = [('scaler', StandardScaler()), ('model', lgb.LGBMRegressor())]
pipeline = Pipeline(steps=steps)
n_splits = 5
cv = KFold(n_splits=n_splits, shuffle=True)
scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='r2')

# Result
print(f"Mean(LGBM Regressor): {np.mean(scores)}")
print(f"std(LGBM Regressor): {np.std(scores)}")
```

## Predictions

```python
# Fitting & Prediction
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
```

```python
# Submission dataframe
sub_data[target_col] = y_pred

# CSV file preparation
sub_data.to_csv('submission.csv', index=False)
```
